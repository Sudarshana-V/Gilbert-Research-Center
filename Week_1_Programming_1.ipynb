```python
#Programming-1 : Number of E-coli bacteria vs relative frequency
import matplotlib.pyplot as plot
import numpy as num
x=num.array([1,2,6,7,8,9,11,12,13,14,15,16,17,18,21,22])
y=[]
for i in x:
  y_val=num.power(2,i)
  y.append(y_val)
plot.hist(y,bins=5,density=True)
plot.title("Distribution Of E.coli Bacteria Growth Under Microscope")
plot.xlabel("Number Of E.coli Bacteria")
plot.ylabel("Relative Frequency")
plot.grid(True, color='pink', linestyle='--', linewidth='0.5')
plot.show()
```
