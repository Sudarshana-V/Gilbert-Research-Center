```python
# Programming-2 : Single-stranded RNA SARS-CoV-2 Covid-19
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plots
import numpy as num

def x(t):
  return num.sqrt(8)*num.sqrt(1 - num.cos(2*t)) / num.sqrt(2)
def y(t):
  return num.sqrt(8)*num.sin(t + num.pi/2)
def z(t):
  return t

t_min = 0
t_max = 3
num_points = 100

t = num.linspace(t_min, t_max, num_points)

x_val = x(t)
y_val = y(t)
z_val = z(t)

fig = plots.figure(figsize=(8, 6))
axes = fig.add_subplot(111, projection='3d')

axes.plot(x_val, y_val, z_val, color='blue', linewidth=2)

axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_zlabel("Z")
axes.set_title("3D Line Plot of SARS CoV-2 RNA Model")

axes.grid(True)
plots.show()
```
