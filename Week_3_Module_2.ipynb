```python
import numpy as num
import matplotlib.pyplot as plot

years = num.array([2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013])
national_waterways_1 = num.array([1596, 3119, 4109, 5803, 7092, 7068, 10402, 12302, 14546, 15119, 18512])
national_waterways_2 = num.array([296, 381, 322, 1728, 422, 142, 74, 573, 613, 580, 594])
national_waterways_3 = num.array([219, 154, 169, 149, 89, 109, 97, 142, 132, 139, 116])

data = num.vstack((national_waterways_1, national_waterways_2, national_waterways_3)).T

centered_data = data - num.mean(data)

covariance_matrix = num.cov(centered_data.T)

eigenvalues, eigenvectors = num.linalg.eig(covariance_matrix)

sorted_index = num.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_index]
eigenvectors = eigenvectors[:, sorted_index]

num_components = 2

projected_data = centered_data.dot(eigenvectors[:, :num_components])

pc1 = projected_data[:, 0]
pc2 = projected_data[:, 1]

# Plotting
plot.scatter(pc1, pc2)
plot.xlabel('Principal Component 1')
plot.ylabel('Principal Component 2')
plot.title('National Waterways Data (PCA)')
plot.legend()
plot.show()
```
